```python
import yfinance as yf 
import pandas as pd
import numpy as np 
import smtplib
from email.mime.text import MIMEText
```

```python
oil = yf.download("BZ=F", period="1y")
oil.head()
```

```python
oil2 = oil['Close']
oil2
```

```python
df = oil2.copy()
df['rolling_40'] = oil2.rolling(40).mean()
df['rolling_90'] = oil2.rolling(90).mean()

df = df.dropna()

print(df)
```

```python
df['signal'] = 0
df['signal'] = np.where(df['rolling_40'] > df['rolling_90'],
                         1,
                         -1
                         )
df['crossover'] = df['signal'].diff()
df = df.dropna()
df.tail()
```

```python
Last_Value = df.iloc[-1]
print(Last_Value)
```

```python
if Last_Value['crossover'] == 2:
    print("Positive increasing trend")
elif Last_Value['crossover'] == -2:
    print("Negative decreasing trend"),
else:
    print("Nessun crossover oggi")
```
